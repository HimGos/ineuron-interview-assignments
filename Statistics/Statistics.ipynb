### Q2. Consider a dataset containing the heights (in centimeters) of 1000 individuals. The mean height is 170 cm with a standard deviation of 10 cm. The dataset is approximately normally distributed, and its skewness is approximately zero. Based on this information, answer the following questions:

- a. What percentage of individuals in the dataset have heights between 160 cm and 180 cm?
- b. If we randomly select 100 individuals from the dataset, what is the probability that their average height is greater than 175 cm?
- c. Assuming the dataset follows a normal distribution, what is the z-score corresponding to a height of 185 cm?
- d. We know that 5% of the dataset has heights below a certain value. What is the approximate height corresponding to this threshold?
- e. Calculate the coefficient of variation (CV) for the dataset.
- f. Calculate the skewness of the dataset and interpret the result.

### Ans 2.

#### a. What percentage of individuals in the dataset have heights between 160 cm and 180 cm?

Given data:
- Mean height ($\mu$) = 170 cm
- Standard deviation ($\sigma$) = 10 cm


We want to calculate $P(160 , \text{cm} < X < 180 , \text{cm})$, which is equivalent to $P\left(\frac{{160 - \mu}}{{\sigma}} < Z < \frac{{180 - \mu}}{{\sigma}}\right)$.

Substituting the values:
$P\left(\frac{{160 - 170}}{{10}} < Z < \frac{{180 - 170}}{{10}}\right)$

Simplifying:
$P(-1 < Z < 1)$

To find this probability, we need to find the area under the standard normal curve between the Z-scores of -1 and 1. The Z-table provides the cumulative probability up to a given Z-score.

Looking up the Z-scores -1 and 1 in the Z-table, we find the following cumulative probabilities:

$P(Z < -1) \approx 0.1587$

$P(Z < 1) \approx 0.8413$

To find the area between -1 and 1, we subtract the probability of Z being less than -1 from the probability of Z being less than 1:

$P(-1 < Z < 1) = P(Z < 1) - P(Z < -1)$

$P(-1 < Z < 1) \approx 0.8413 - 0.1587 = 0.6827$

=> **68.27%** 

#### b. If we randomly select 100 individuals from the dataset, what is the probability that their average height is greater than 175 cm?

Mean population height($\mu$) = 170 cm

Standard deviation ($\sigma$) = 10 cm

Sample size ($n$) = 100 individuals

Standard error of the sample mean (SE) = $\frac{\sigma}{\sqrt{n}}$

SE = $\frac{10}{\sqrt{100}} = 1$

Now, we need to calculate the Z-score for the sample mean of 175 cm:

Z = $\frac{\bar{X} - \mu}{SE}$

Here, $\bar{X}$ is the sample mean, which in this case is 175 cm.

Z = $\frac{175 - 170}{1} = 5$

To find the probability of a Z-score greater than 5, we can use the Z-table or a calculator. However, since the Z-score of 5 is significantly large, it is unlikely to have a precise value in the Z-table.

Instead, we can use an approximation and assume that the probability of a Z-score greater than 5 is very close to 0. Therefore, the probability that the average height of a randomly selected sample of 100 individuals is greater than 175 cm can be considered as effectively 0, or extremely close to 0.

#### c. Assuming the dataset follows a normal distribution, what is the z-score corresponding to a height of 185 cm?

Mean height ($\mu$) = 170 cm

Standard deviation ($\sigma$) = 10 cm

Height ($X$) = 185 cm

We will use formula:
$Z = \frac{{X} - \mu}{\sigma}$

$Z = \frac{{185} - 170}{10} = 1.5$

#### d. We know that 5% of the dataset has heights below a certain value. What is the approximate height corresponding to this threshold?

We need to find the Z-score that corresponds to the cumulative probability of 5% (0.05).

Z-score value of 0.05 = -1.645

$\mu$ = 170 cm

$\sigma$ = 10 cm

$X = -1.645 * 10 + 170$

$X \approx153.55$

Approximate height is **153.55cm**.

#### e. Calculate the coefficient of variation (CV) for the dataset.

$\mu$ = 170 cm

$\sigma$ = 10 cm

Calculating CV =  ($\frac{\sigma}{\mu}$) * 100

CV = ($\frac{10}{170}$) * 100

CV = 5.88

CV of the dataset is approx **5.88%**

### Q3. Consider the ‘Blood Pressure Before’ and ‘Blood Pressure After’ columns from the data and calculate the following:
    
a. Measure the dispersion in both and interpret the results.

b. Calculate mean and 5% confidence interval and plot it in a graph

c. Calculate the Mean absolute deviation and Standard deviation and interpret the results.

d. Calculate the correlation coefficient and check the significance of it at 1% level of significance.

### Ans 3.

In our dataset, we have record of 100 patients with 2 rows... `Blood Pressure Before` and `Blood Pressure After`.

#### a. Measure the dispersion in both and interpret the results.

```python
import pandas as pd

bp = pd.read_csv('data.csv')

bp.head()
```

```python
before_dispersion = {
    'Range': bp[' Blood Pressure Before (mmHg)'].max() - bp[' Blood Pressure Before (mmHg)'].min(),
    'Variance': bp[' Blood Pressure Before (mmHg)'].var(),
    'Standard Deviation': bp[' Blood Pressure Before (mmHg)'].std()
}

after_dispersion = {
    'Range': bp[' Blood Pressure After (mmHg)'].max() - bp[' Blood Pressure After (mmHg)'].min(),
    'Variance': bp[' Blood Pressure After (mmHg)'].var(),
    'Standard Deviation': bp[' Blood Pressure After (mmHg)'].std()
}

print("Dispersion in Blood Pressure Before:-")
for key, value in before_dispersion.items():
    print(f"{key}: {round(value,2)}")
    
print("\nDispersion in Blood Pressure After:-")
for key, value in after_dispersion.items():
    print(f"{key}: {round(value,2)}")
```

#### b. Calculate mean and 5% confidence interval and plot it in a graph

```python
import matplotlib.pyplot as plt
from scipy import stats

# Blood Pressure Before mean
bp_before_mean = bp[' Blood Pressure Before (mmHg)'].mean()

# Blood Pressure After mean
bp_after_mean = bp[' Blood Pressure After (mmHg)'].mean()

# Calculate 5% confidence interval for 'Blood Pressure Before' column
bp_before_ci = stats.t.interval(0.95, # CI
                                len(bp[' Blood Pressure Before (mmHg)']) - 1, 
                                loc=bp_before_mean, 
                                scale=stats.sem(bp[' Blood Pressure Before (mmHg)']))

# Calculate 5% confidence interval for 'Blood Pressure After' column
bp_after_ci = stats.t.interval(0.95,  # CI
                               len(bp[' Blood Pressure After (mmHg)']) - 1, 
                               loc=bp_after_mean, 
                               scale=stats.sem(bp[' Blood Pressure After (mmHg)']))

# Plot the means and confidence intervals in a graph
x = ['Blood Pressure Before', 'Blood Pressure After']
y = [bp_before_mean, bp_after_mean]
y_err = [bp_before_ci[1] - bp_before_mean, bp_after_ci[1] - bp_after_mean]

plt.bar(x, y, yerr=y_err, capsize=5)
plt.xlabel('Blood Pressure')
plt.ylabel('Mean')
plt.title('Mean with 5% Confidence Interval')
plt.show();
```

#### c. Calculate the Mean absolute deviation and Standard deviation and interpret the results.

```python
before_mad_std = {
    'Mean Absolute Deviation': (bp[' Blood Pressure Before (mmHg)'] - bp[' Blood Pressure Before (mmHg)'].mean()).abs().mean(),
    'Standard Deviation': bp[' Blood Pressure Before (mmHg)'].std()
}

after_mad_std = {
    'Mean Absolute Deviation': (bp[' Blood Pressure After (mmHg)'] - bp[' Blood Pressure After (mmHg)'].mean()).abs().mean(),
    'Standard Deviation': bp[' Blood Pressure After (mmHg)'].std()
}

print("Mean Absolute Deviation & Standard Deviation of BP Before")
for key, value in before_mad_std.items():
    print(f"{key}: {value}")
    
print("\nMean Absolute Deviation & Standard Deviation of BP After")
for key, value in after_mad_std.items():
    print(f"{key}: {value}")
```

#### d. Calculate the correlation coefficient and check the significance of it at 1% level of significance.

```python
# Calculate the correlation coefficient and p-value
correlation_coefficient, p_value = stats.pearsonr(bp[' Blood Pressure Before (mmHg)'], bp[' Blood Pressure After (mmHg)'])

# Print the correlation coefficient and p-value
print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)

# Check significance at 1% level of significance
alpha = 0.01

if p_value < alpha:
    print("The correlation is significant at the 1% level of significance.")
else:
    print("The correlation is not significant at the 1% level of significance.")
```

### Q5. A certain city has two taxi companies: Company A has 80% of the taxis and Company B has 20% of the taxis. Company A's taxis have a 95% success rate for picking up passengers on time, while Company B's taxis have a 90% success rate. If a randomly selected taxi is late, what is the probability that it belongs to Company A?

### Ans 5.

Let's assume there are 100 taxis.

80 Taxis belong to A and 20 Taxis belong to B. 

Now, Late taxis from A = $\frac{5}{100} . 80$ = 4

Late taxis from B = $\frac{10}{100} . 20$ = 2

Total late taxis = 6

Chances of late taxi to belong to A = $\frac{4}{6}$ = **66.7%**

---

We can solve it using Bayes's theorem too:

- P(A) = 0.8   (Company A)
- P(B) = 0.2   (Company B)
- P(L|A) = 0.05 (Late taxi belong to A)
- P(L|B) = 0.10 (Late taxi belong to B)

P(L) = P(A) * P(L|A) + P(B) * P(L|B) = (0.8 * 0.05) + (0.2 * 0.1)

P(L) = 0.04 + 0.02 = 0.06  

P(A|L) = (P(A) * P(L|A)) / P(L) = (0.8 * 0.05) / 0.06

P(A|L) = 4/6 = 0.667

Therefore, the probability that late taxi belongs to Company A is **66.7%**

### Q6. A pharmaceutical company is developing a drug that is supposed to reduce blood pressure. They conduct a clinical trial with 100 patients and record their blood pressure before and after taking the drug. The company wants to know if the change in blood pressure follows a normal distribution.

### Ans 6.

```python
bp = pd.read_csv('data.csv')

bp.head()
```

```python
from scipy.stats import shapiro

# Calculate the change in blood pressure
bp['Change in Blood Pressure'] = bp[' Blood Pressure After (mmHg)'] - bp[' Blood Pressure Before (mmHg)']

# Perform Shapiro-Wilk test
stat, p_value = shapiro(bp['Change in Blood Pressure'])

# Check if p-value is less than the significance level (e.g., 0.05) to reject the null hypothesis
significance_level = 0.05
if p_value < significance_level:
    print("The change in blood pressure does not follow a normal distribution.")
else:
    print("The change in blood pressure follows a normal distribution.")
```

### Q8. The anxiety levels of 10 participants were measured before and after a new therapy. The scores are not normally distributed. Use the Wilcoxon signed-rank test to test whether the therapy had a significant effect on anxiety levels. The data is given below: Participant Before therapy & After therapy Difference

```python
participant = [i for i in range(1,11)]
before = [10, 8, 12, 15, 6, 9, 11, 7, 14, 10]
after = [7, 6, 10, 12, 5, 8, 9, 6, 12, 8]
diff = [-3, -2, -2, -3, -1, -1, -2, -1, -2, -2]
```

### Ans 8.

```python
from scipy.stats import wilcoxon

# Extract the before and after scores from the dataframe
before_scores = df['Before Therapy']
after_scores = df['After Therapy']

# Perform Wilcoxon signed-rank test
statistic, p_value = wilcoxon(before_scores, after_scores)

# Print the test statistic and p-value
print("Test Statistic:", statistic)
print("p-value:", p_value)
```

> The test statistic is 0.0, and the p-value is approximately 0.0019. Since the p-value is less than the conventional significance level of 0.05, we reject the null hypothesis. This indicates that there is sufficient evidence to conclude that the therapy had a significant effect on anxiety levels.

### Q9. Test the hypothesis that the mean scores of all the students are the same. If not, name the student with the highest score.

```python
name = ['Karan', 'Deepa', 'Karthik', 'Chandan', 'Jeevan']
exam1 = [85, 70, 90, 75, 95]
exam2 = [90, 80, 85, 70, 92]
final = [92, 85, 88, 75, 96]

data = {'Name': name, 'Exam 1': exam1, 'Exam 2': exam2, 'Final Exam': final}

df = pd.DataFrame(data)
df
```

### Ans 9.

To test the hypothesis that the mean scores of all the students are the same, we can use a one-way **analysis of variance (ANOVA)** test. This test compares the means of multiple groups to determine if there are statistically significant differences among them. In this case, we will compare the mean scores of the students from the three exams.

```python
from scipy.stats import f_oneway

# Extract the exam scores from the dataframe
exam1_scores = df['Exam 1']
exam2_scores = df['Exam 2']
final_scores = df['Final Exam']

# Perform one-way ANOVA test
f_statistic, p_value = f_oneway(exam1_scores, exam2_scores, final_scores)

# Print the F-statistic and p-value
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Find the student with the highest score
highest_score_student = df.loc[df['Final Exam'].idxmax(), 'Name']

# Print the student with the highest score
print("Student with the highest score:", highest_score_student)
```

> The corrected F-statistic is approximately 0.3237, and the p-value is approximately 0.7296. Since the p-value is greater than the conventional significance level of 0.05, we fail to reject the null hypothesis. This indicates that there is no significant difference in the mean scores of all the students. 

The student with the highest score is Jeevan.

### Q10. A factory produces light bulbs, and the probability of a bulb being defective is 0.05.
### The factory produces a large batch of 500 light bulbs.
### a. What is the probability that exactly 20 bulbs are defective?
### b. What is the probability that at least 10 bulbs are defective?
### c. What is the probability that at max 15 bulbs are defective?
### d. On average, how many defective bulbs would you expect in a batch of 500?

### Ans 10.

Using Binomial Distribution: 
    
    P(X = k) = C(n, k) * p^k * (1 - p)^(n - k)

```python
from scipy.stats import binom

# Parameters
n = 500  # Number of trials (bulbs)
p = 0.05  # Probability of success (bulb being defective)

# a. Probability of exactly 20 bulbs being defective
k1 = 20  # Number of successful trials
prob_20_defective = binom.pmf(k1, n, p)

# b. Probability of at least 10 bulbs being defective
k2 = range(10, n+1)
prob_at_least_10_defective = sum(binom.pmf(k, n, p) for k in k2)

# c. Probability of at most 15 bulbs being defective
k3 = range(16)
prob_at_most_15_defective = sum(binom.pmf(k, n, p) for k in k3)

# d. Average number of defective bulbs
mean_defective_bulbs = n * p

# Print the results
print(f"a. Probability of exactly 20 bulbs being defective: {prob_20_defective:.2f}")
print(f"b. Probability of at least 10 bulbs being defective: {prob_at_least_10_defective:.3f}")
print(f"c. Probability of at most 15 bulbs being defective: {prob_at_most_15_defective:.3f}")
print(f"d. Average number of defective bulbs: {mean_defective_bulbs:.3f}")
```

```python
from scipy.stats import bartlett

# Perform Bartlett's test
_, p_value = bartlett(before_values, after_values)

# Check if p-value is significant
alpha = 0.05
if p_value < alpha:
    print("The variances of the classes are significantly different.")
else:
    print("The variances of the classes are not significantly different.")
```

### Q12. A pharmaceutical company develops a new drug and wants to compare its effectiveness against a standard drug for treating a particular condition. They conduct a study with two groups: 

### Group A receives the new drug, and Group B receives the standard drug. The company measures the improvement in a specific symptom for both groups after a 4-week treatment period.

#### a. The company collects data from 30 patients in each group and calculates the mean improvement score and the standard deviation of improvement for each group. The mean improvement score for Group A is 2.5 with a standard deviation of 0.8, while the mean improvement score for Group B is 2.2 with a standard deviation of 0.6. Conduct a t-test to determine if there is a significant difference in the mean improvement scores between the two groups. Use a significance level of 0.05.

#### b. Based on the t-test results, state whether the null hypothesis should be rejected or not. Provide a conclusion in the context of the study.

### Ans 12. 

**a.**


```python
import scipy.stats as stats

# Define the data for each group
mean_a = 2.5
std_a = 0.8
n_a = 30

mean_b = 2.2
std_b = 0.6
n_b = 30

# Perform independent two-sample t-test
t_stat, p_value = stats.ttest_ind_from_stats(mean_a, std_a, n_a, mean_b, std_b, n_b)

# Check if p-value is significant
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in the mean improvement scores between the two groups.")
else:
    print("There is no significant difference in the mean improvement scores between the two groups.")
```

**b.**

Based on the t-test results, we evaluate the null hypothesis. In this case, the null hypothesis states that there is no significant difference in the mean improvement scores between Group A and Group B.

If the p-value obtained from the t-test is less than the significance level (0.05 in this case), we reject the null hypothesis. This would indicate that there is a significant difference in the mean improvement scores between the two groups.

If the p-value is greater than or equal to the significance level, we fail to reject the null hypothesis. This would suggest that there is no significant difference in the mean improvement scores between the two groups.

Therefore, in the context of the study, if the t-test yields a p-value less than 0.05, we can conclude that there is a significant difference in the mean improvement scores between the two groups. On the other hand, if the p-value is greater than or equal to 0.05, we would conclude that there is no significant difference in the mean improvement scores between Group A and Group B.
