## Imagine you working as a sales manager now you need to predict the `Revenue` and whether that particular revenue is on the `weekend` or not and find the `Informational_Duration` using the Ensemble learning algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

### Model Training

```python
df_new = df.copy()
```

```python
# Converting weekend olumn into numerical

df_new['Weekend'] = df_new['Weekend'].astype(int)

df_new['Weekend'][:5]
```

```python
# Converting informational_duration column into numerical
df_new["Informational_Duration"] = df_new['Informational_Duration'].apply(lambda x: 1 if x > 0 else 0)

df_new.head()
```

```python
df_new['Revenue'] = df_new['Revenue'].astype(str)
```

```python
features = df_new.drop(['Revenue', 'Weekend', 'Informational_Duration'], axis=1)

# Encode categorical features using one-hot encoding
features = pd.get_dummies(features)

y_revenue = df_new['Revenue']
y_weekend = df_new['Weekend']
y_info = df_new['Informational_Duration']

# Splitting data
X_train, X_test, y_revenue_train, y_revenue_test, weekend_train, weekend_test, info_duration_train, info_duration_test = train_test_split(
    features, y_revenue, y_weekend, y_info, test_size=0.2, random_state=42)
```

```python
# Training the ensemble model (Random Forest classifier) for revenue prediction
revenue_model = RandomForestClassifier(n_estimators=100)
revenue_model.fit(X_train, y_revenue_train)

# Making predictions on the revenue
revenue_predictions = revenue_model.predict(X_test)

# Assessing accuracy of the revenue predictions
revenue_accuracy = accuracy_score(y_revenue_test, revenue_predictions)
print("Accuracy for revenue prediction:", revenue_accuracy)
```

```python
# Training the ensemble model (Random Forest classifier) for weekend prediction
weekend_model = RandomForestClassifier(n_estimators=100)
weekend_model.fit(X_train, weekend_train)

# Making predictions on the weekend
weekend_predictions = weekend_model.predict(X_test)

# Assessing accuracy of the weekend predictions
weekend_accuracy = accuracy_score(weekend_test, weekend_predictions)
print("Accuracy for weekend prediction:", weekend_accuracy)
```

```python
# Training the ensemble model (Random Forest classifier) for INformational_duration prediction
info_model = RandomForestClassifier(n_estimators=100)
info_model.fit(X_train, info_duration_train)

# Making predictions on the weekend
info_predictions = info_model.predict(X_test)

# Assessing accuracy of the weekend predictions
info_accuracy = accuracy_score(info_duration_test, info_predictions)
print("Accuracy for informatioal duration prediction:", info_accuracy)
```
